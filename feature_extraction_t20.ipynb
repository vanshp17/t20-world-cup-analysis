```python
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pickle.load(open('t20_dataset_l2.pkl','rb'))
```

```python
df

'''
    atting team
    bowling team
    city
    current score
    wickets left
    ball left
    current rr
    last five
'''
```

```python
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
import numpy as np
```

```python
cities = np.where(df['city'].isnull(), 
         df['venue'].str.split().apply(lambda x:x[0]),
        df['city'])
```

```python
df['city'] = cities
```

```python
df.isnull().sum()
```

```python
df.drop(columns='venue',inplace=True)
```

```python
df
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df.sample()
```

```python
df['current_score'] =df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
#balls_left

df['over'] = df['ball'].apply(lambda x: str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x: str(x).split(".")[1])
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x: 0 if x<0 else x)
```

```python
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x: 0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
# current runrate

df['crr'] = (df['current_score']*6)/ df['balls_bowled']
```

```python
df
```

```python
df.isnull().sum()
```

```python
#last five overs run

groups = df.groupby('match_id')
match_ids = df['match_id'].unique()
last_five = []
# Calculate runs in the last five overs (30 balls)
for id in match_ids:
    group = groups.get_group(id)
    last_30_balls_runs = group['runs'].rolling(window=30).sum().dropna().values.tolist()
    last_five.extend(last_30_balls_runs)
```

```python
df['last_five'] = last_five
```

```python
groups = df.groupby('match_id')
df['last_five_overs_runs'] = 0

# Calculate runs in the last five overs (30 balls) for each match
for id, group in groups:
    group_size = len(group)
    if group_size >= 30:
        df.loc[group.index, 'last_five_overs_runs'] = group['runs'].rolling(window=30).sum()
    else:
        df.loc[group.index, 'last_five_overs_runs'] = None
```

```python
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df, on='match_id')
```

```python
final_df = final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five_overs_runs','runs_x']]
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
#to remove any type bias
final_df  = final_df.sample(final_df.shape[0])
```

```python
X = final_df.drop(columns = 'runs_x')
y = final_df['runs_x']
```

```python
X_train,X_test,y_train, y_test = train_test_split(X,y, test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
final_df.columns
```

```python
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False,
                         drop='first'),
     ['batting_team', 'bowling_team', 'city'])
],
remainder='passthrough')
```

```python
pipe = Pipeline(steps = [
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12,
                         random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)

print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('t20_pipe.pkl','wb'))
```
